```python
#prompt template text summarization
from langchain.chains import LLMChain
from langchain import PromptTemplate
GenericpromptTemplate="""
Write a summary of the following speech:
Speech:{speech}
Translate the precise summary to {language}
"""

#in order to convert this as prompt we will use prompt template
prompt=PromptTemplate(
    input_variable=['speech','language'],
    template=GenericpromptTemplate
)

```

```python
complete_prompt=prompt.format(speech=speech,language="French")
complete_prompt
```

```python
llm.get_num_tokens(complete_prompt)
```

```python
llm_chain=LLMChain(llm=llm,prompt=prompt)
summary=llm_chain.run({'speech':speech,'language':'hindi'})
summary
```

```python
#when we have the bigger PDF 
#stuff document chains 
from langchain_community.document_loaders import PyPDFLoader
loader=PyPDFLoader(r"C:\Users\hp\Desktop\Summarization GenAI\apjspeech.pdf")
docs=loader.load_and_split()
docs

```

```python
#now we will use chunk document to get our work done 
template="""Write a concise and short summary of the following speech,
Speech :{text}

"""

prompt=PromptTemplate(input_variable=['text'],template=template)
```

```python
from langchain.chains.summarize import load_summarize_chain
chain=load_summarize_chain(llm=llm,chain_type='stuff',prompt=prompt,verbose=True)
summary=chain.run(docs)
summary
```

```python
llm.get_num_tokens(summary)
```

```python
#now here we have divided our data into many more chunks so that we can use map reduced technique
from langchain_text_splitters import RecursiveCharacterTextSplitter
final_document=RecursiveCharacterTextSplitter(chunk_size=2000,chunk_overlap=100).split_documents(docs)
```

```python
#what will map reduced technique will do it will take the chunk and and summarize the each chunk and at the end when all chunks has been done it will make final answer with the help of prompt and passed to llm to crrate the all over answers
len(final_document)
```

```python
#now we will create prompt for this chunks 
chunks_prompts="""
please summarize the below speech:
Speech:`{text}`
Summary:



"""
map_prompt_template=PromptTemplate(input_variable=['text'],template=chunks_prompts)
```

```python
#we need to make the final prompt 
final_prompt='''
Provide the final summary of the entire speedch with these important points
Add a motivational title,start the precise summary with an introduction and provide the summary
in number points for the speech.
Speech:{text}

'''
final_prompt_template=PromptTemplate(input_variable=['text'],template=final_prompt)
```

```python
#for all the chunks we will use the abive prompt 
summary_chain=load_summarize_chain(
    llm=llm,
    chain_type='map_reduce',
    map_prompt=map_prompt_template,
    combine_prompt=final_prompt_template,
    verbose=True
)

output=summary_chain.run(final_document)
output
```

```python
#if we need to refine more our data we will use chain_type=refine
summary_chain1=load_summarize_chain(
    llm=llm,
    chain_type='refine',
    verbose=True
)

output=summary_chain1.run(final_document)
output
```
